```python
from duckduckgo_search import ddg_images
from fastcore.all import *

# Function : encuentra max_images imagenes en DuckDuckGo por term | Retorna : lista de urls
def search_images(term, max_images = 200) : return L(ddg_images(term, max_results = max_images)).itemgot('image')
```

```python
urls = search_images('bird picture', max_images = 1) 
urls[0]
```

```python
# Descargar imagen desde una url
from fastdownload import download_url
dest = 'bird.jpg'
download_url(urls[0], dest, show_progress = False)
```

```python
from fastai.vision.all import *
bird = Image.open(dest)
bird.to_thumb(256,256)
```

```python
download_url(search_images('forest picture', max_images=1)[0], 'forest.jpg', show_progress=False)
Image.open('forest.jpg').to_thumb(256,256)
```
